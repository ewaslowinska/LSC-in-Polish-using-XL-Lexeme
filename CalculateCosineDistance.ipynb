```python
from scipy.spatial.distance import cosine
import pickle
```

```python
def cosine_distance(input, input1):
    distances = {}
    for root, dirs, files in os.walk(input):
        for file_name in files:
            file_path = os.path.join(root, file_name)
            matching = '{}/{}'.format(input1, file_name)
            with open(file_path, 'rb') as f:
                c1 = pickle.load(f)
            with open(matching, 'rb') as f1:
                c2 = pickle.load(f1)
            distance = cosine(c1, c2)
            distances[os.path.splitext(file_name)[0]] = distance
    return distances
```
